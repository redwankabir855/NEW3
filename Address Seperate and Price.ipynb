```python
# Create a new dataframe with the average price for each postal code
a_grouped = a.groupby(['Postal Code','Province'])[' Price '].mean().reset_index()

# Rename the columns
a_grouped.columns = ['Postal Code','Province','Avg Price']
```

```python
# Create a new dataframe with the average price for each postal code
a_grouped = a.groupby(['Postal Code','Province'])['Price'].mean().reset_index()

# Rename the columns
a_grouped.columns = ['Postal Code','Province','Avg Price']
```

```python
b['Avg Price'] = b['Avg Price'].astype(float) # Convert the values to float

# Round the values to the nearest integer
b['Avg Price'] = b['Avg Price'].apply(lambda x: round(x))

# Check the return value of round before assigning it to the column
b['Avg Price'] = b['Avg Price'].apply(lambda x: round(x) if round(x) is not None else x)
```
